```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
import torch
import os
os.environ["TRANSFORMERS_NO_TF"] = "1"

# 1. Load base model architecture
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-base")

state_dict = torch.load("t5_finetuned_best.pt")
model.load_state_dict(state_dict)
# 2. Load tokenizer
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")

# 3. Load your fine-tuned weights
# model.load_state_dict(torch.load("t5_finetuned/t5_finetuned_best.pt"))
model.save_pretrained("./fine_tuned_base/")
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")
tokenizer.save_pretrained("./fine_tuned_base/")

# 4. Set model to eval mode
# model.eval()
```
